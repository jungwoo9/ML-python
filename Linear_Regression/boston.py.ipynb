<h1>Data Load</h1>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from sklearn.datasets import load_boston
dataset = load_boston()

print(dataset.data)
```

```python
print(dataset.feature_names)
```

```python
print(dataset.DESCR)
```

```python
print(dataset.target)
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['MEDV'] = dataset.target
df.head()
```

<h3>Visualize Data</h3>

```python
plt.scatter(df['LSTAT'], df['MEDV'], marker='o')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
```

```python
plt.scatter(df['RM'], df['MEDV'], marker='o')
plt.xlabel('RM')
plt.ylabel('MEDV')
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(18,15))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['LSTAT'], df['RM'], df['MEDV'], c='b')
ax.set_xlabel("LSTAT")
ax.set_ylabel("RM")
ax.set_zlabel("MEDV")
plt.show()
```

<h3>Training the Model</h3>

```python
x = pd.DataFrame(np.c_[df['LSTAT'], df['RM']], columns = ['LSTAT', 'RM'])
Y = df['MEDV']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, Y_train, Y_test = train_test_split(x, Y, test_size=0.3, random_state=5)
print(x_train.shape)
print(Y_train.shape)
```

```python
# fit model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, Y_train)
```

```python
# predict price
price_pred = model.predict(x_test)

print('R-squared: %.2f' %model.score(x_test, Y_test))
```

```python
# plot the predicted price
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(Y_test, price_pred)
print(mse)

plt.scatter(Y_test, price_pred)
plt.title('Actual prices vs Predicted prices')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted prices')
plt.grid()
plt.show()
```

<h3>Y = a<sub>0</sub> + a<sub>1</sub>x<sub>1</sub> + a<sub>2</sub>x<sub>2</sub></h3>
<br>Y = dependent variable
<br>a<sub>0</sub> = intercept
<br>a<sub>1</sub>, a<sub>2</sub> = coefficient
<br>x<sub>1</sub>, x<sub>2</sub> = features

```python
# intercept
print(model.intercept_)

# coefficient
print(model.coef_)
```

<h3>Plot 3D hyperplane</h3>

```python
fig = plt.figure(figsize=(18,15))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x['LSTAT'], x['RM'], Y, c='b')

ax.set_xlabel("LSTAT")
ax.set_ylabel("RM")
ax.set_zlabel("MEDV")

# create a meshgrid of all the values for LSTAT and RM
x_surf = np.arange(0, 40, 1)
y_surf = np.arange(0, 10, 1)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)

model = LinearRegression()
model.fit(x, Y)

z = lambda x, y: (model.intercept_ + model.coef_[0] * x + model.coef_[1] * y)

ax.plot_surface(x_surf, y_surf, z(x_surf, y_surf), rstride=1, cstride=1, color='None', alpha=0.4)
plt.show()
```
